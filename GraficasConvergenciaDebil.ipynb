```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
x= np.arange(9)
u= np.ones_like(x)
y= [5, 4.5, 4.5, 6.8, 1.4, 6, 4, 3.3, 2]
t=0
n=1
C=10
#plotting los datos
plt.xlabel('Partición uniforme de [0,T]')
plt.ylabel('$\widetilde{u}_{k,h}(a_j,t)$')
plt.title('Evolución temporal de $\widetilde{u}_{k,h}$')
plt.grid()#poner grilla
plt.xticks(x,[r'$0$',r'$t_1$',r'$t_2$',r'$t_3$',r'...',r'$t_{n-1}$',r'$t_n$',r'...', r'T'],fontsize=14)
plt.yticks(fontsize=14)
#plt.yticks([-C, 0, +C],[r'$-C$', r'$0$', r'$+C$'])
#plt.legend(frameon=False)
plt.plot(x,y,color='blue')
plt.plot(x,y, "o", color='blue',markersize=4)
axes = plt.gca()
axes.title.set_size(20)
axes.xaxis.label.set_size(16)
axes.yaxis.label.set_size(16)
#axes.xticks.label.set_size(14)
#axes.yticks.label.set_size(14)
plt.savefig('1.pdf', dpi=300, bbox_inches='tight')
```
